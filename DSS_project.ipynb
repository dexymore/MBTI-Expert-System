```python


class PersonalityExpert(KnowledgeEngine):
    # @DefFacts()
    def __init__(self):
        super().__init__()
 
        self.answers = []
       
      
        self.personality=None 
        self._initial_action()

    def calculate_majority(self, facts):
        count_A = sum(1 for f in facts if f[1] == 'A')
        count_B = sum(1 for f in facts if f[1] == 'B')
        return 'A' if count_A > count_B else 'B'

    @DefFacts()
    def _initial_action(self):
        def show_questions_set1():
# Remove welcome screen
            welcome_frame.pack_forget()
            
            # Show questions screen   
            header_label.pack(fill='x')
            questionsSet1_frame.pack(expand=True)


        def show_questions_set2():
            # Remove welcome screen
            questionsSet1_frame.pack_forget()

            # Show questions screen
            questionsSet2_frame.pack(expand=True)
            header_label.pack(fill='x')
        def show_questions_set3():
            # Remove welcome screen
            questionsSet2_frame.pack_forget()
            # Show questions screen
            questionsSet3_frame.pack(expand=True)
            header_label.pack(fill='x')

            # Remove questions screen





        

        def collect_answers():
            # Collect all answers from each set of questions
            self.answers = [
                extraverted.get(),   # Set 1
                day_preference.get(),
                energy_source.get(),
                focus.get(),
                concrete_abstract.get(),  # Set 2
                experience_imagination.get(),
                logic_emotions.get(),
                fairness_efficiency.get(),
                expressing_feelings.get(),  # Set 3
                structured_lifestyle.get(),
                quick_decisions_options.get(),
                closure_possibilities.get()
]
        
            # self.determine_personality()
     
            self.form.destroy()
       
            
     
 
       
    

       
    
          

            
        self.form = Tk()
        self.form.title("Personality Type Expert System")
        self.form.geometry("800x600")
        self.form.configure(bg='#fffff6')
        header_label = Label( self.form, text="Personality Type Expert System", font=("Arial Bold", 16), bg='black', fg='white', pady=14, padx=14, anchor='w')


        header_label.pack_forget()

        welcome_frame = Frame( self.form, bg='#fffff6')
        welcome_frame.pack(expand=True)

        # Welcome message label
        lb1 = Label(welcome_frame, text="Welcome to the Personality Type Expert System!", font=("Arial Bold", 18), bg='#fffff6')
        lb1.pack(pady=20)

        # Instruction label
        lb2 = Label(welcome_frame, text="Please answer the following questions to determine your personality type.", font=("Arial Bold", 10), bg='#fffff6')
        lb2.pack(pady=20)

        # Start button
        try:
            startIcon = PhotoImage(file="start.png")
            btn_start = Button(welcome_frame, text="Start now", image=startIcon,  compound="right", bg='black', fg='white', font=("Arial Bold", 16), padx=20, pady=10, command=show_questions_set1)
            btn_start.image = startIcon  # Keep a reference!
        except TclError:
            print("Image 'start.png' failed to load. Using text button instead.")
            btn_start = Button(welcome_frame, text="Start now", bg='black', fg='white', font=("Arial Bold", 16), padx=20, pady=10, command=show_questions_set1)
        btn_start.pack(pady=20)

        # Frame for questions
        questionsSet1_frame = Frame( self.form, bg='#fffff6')
        questionsSet2_frame = Frame( self.form, bg='#fffff6')
        questionsSet3_frame = Frame( self.form, bg='#fffff6')
        results_frame = Frame( self.form, bg='#fffff6')
      
        

        questionsSet1_frame.pack(side=LEFT, fill=BOTH, expand=True, anchor='w')  # Anchor to the left
        questionsSet2_frame.pack(side=LEFT, fill=BOTH, expand=True, anchor='w')  # Anchor to the left
        questionsSet3_frame.pack(side=LEFT, fill=BOTH, expand=True, anchor='w') 
        results_frame.pack(side=LEFT, fill=BOTH, expand=True, anchor='w')  # Anchor to the left




        # Variables for radio buttons
        extraverted = StringVar(value="A")
        day_preference = StringVar(value="A")
        energy_source = StringVar(value="A")
        focus = StringVar(value="A")
        concrete_abstract = StringVar(value="A")
        experience_imagination = StringVar(value="A")
        logic_emotions = StringVar(value="A")
        fairness_efficiency = StringVar(value="A")
        expressing_feelings = StringVar(value="A")
        structured_lifestyle = StringVar(value="A")
        quick_decisions_options = StringVar(value="A")
        closure_possibilities = StringVar(value="A")



        # Question 1
        Label(questionsSet1_frame, text="When do you enjoy the most:", font=("Arial Bold", 18), bg='#fffff6').pack(pady=10, anchor='w')
        Radiobutton(questionsSet1_frame, text="A) Social gatherings and being around people", variable=extraverted, value="A", bg='#fffff6').pack(anchor='w')
        Radiobutton(questionsSet1_frame, text="B) Prefer to spend quiet time alone or with a few close friends", variable=extraverted, value="B", bg='#fffff6').pack(anchor='w')

        # Question 2
        Label(questionsSet1_frame, text="After a long day, would you rather:", font=("Arial Bold", 18), bg='#fffff6').pack(pady=10, anchor='w')
        Radiobutton(questionsSet1_frame, text="A) Go to a party", variable=day_preference, value="A", bg='#fffff6').pack(anchor='w')
        Radiobutton(questionsSet1_frame, text="B) Stay home and read a book", variable=day_preference, value="B", bg='#fffff6').pack(anchor='w')

        # Question 3
        Label(questionsSet1_frame, text="When do you feel energized:", font=("Arial Bold", 18), bg='#fffff6').pack(pady=10, anchor='w')
        Radiobutton(questionsSet1_frame, text="A) By interacting with others", variable=energy_source, value="A", bg='#fffff6').pack(anchor='w')
        Radiobutton(questionsSet1_frame, text="B) By spending time alone", variable=energy_source, value="B", bg='#fffff6').pack(anchor='w')

        # Question 4
        Label(questionsSet1_frame, text="Are you more focused on:", font=("Arial Bold", 18), bg='#fffff6').pack(pady=10, anchor='w')
        Radiobutton(questionsSet1_frame, text="A) Facts and details", variable=focus, value="A", bg='#fffff6').pack(anchor='w')
        Radiobutton(questionsSet1_frame, text="B) Possibilities and the big picture", variable=focus, value="B", bg='#fffff6').pack(anchor='w')

        # Set 2 questions
        Label(questionsSet2_frame, text="Do you prefer working:", font=("Arial Bold", 18), bg='#fffff6').pack(pady=10, anchor='w')
        Radiobutton(questionsSet2_frame, text="A) With tangible, concrete in self.formation.", variable=concrete_abstract, value="A", bg='#fffff6').pack(anchor='w')
        Radiobutton(questionsSet2_frame, text="B) Abstract theories and ideas.", variable=concrete_abstract, value="B", bg='#fffff6').pack(anchor='w')

        Label(questionsSet2_frame, text="When solving problems, do you rely more on:", font=("Arial Bold", 18), bg='#fffff6').pack(pady=10, anchor='w')
        Radiobutton(questionsSet2_frame, text="A) Your past experiences", variable=experience_imagination, value="A", bg='#fffff6').pack(anchor='w')
        Radiobutton(questionsSet2_frame, text="B) Your imagination", variable=experience_imagination, value="B", bg='#fffff6').pack(anchor='w')

        Label(questionsSet2_frame, text="Are you more likely to make decisions based on:", font=("Arial Bold", 18), bg='#fffff6').pack(pady=10, anchor='w')
        Radiobutton(questionsSet2_frame, text="A) Logic and reason", variable=logic_emotions, value="A", bg='#fffff6').pack(anchor='w')
        Radiobutton(questionsSet2_frame, text="B) Values and emotions", variable=logic_emotions, value="B", bg='#fffff6').pack(anchor='w')

        Label(questionsSet2_frame, text="Do you prioritize in decision-making:", font=("Arial Bold", 18), bg='#fffff6').pack(pady=10, anchor='w')
        Radiobutton(questionsSet2_frame, text="A) Fairness and justice", variable=fairness_efficiency, value="A", bg='#fffff6').pack(anchor='w')
        Radiobutton(questionsSet2_frame, text="B) Efficiency and objectivity", variable=fairness_efficiency, value="B", bg='#fffff6').pack(anchor='w')

        # Set 3 questions
        # Set 3 questions
        Label(questionsSet3_frame, text="Are you more comfortable expressing:", font=("Arial Bold", 18), bg='#fffff6').pack(pady=10, anchor='w')
        Radiobutton(questionsSet3_frame, text="A) Your thoughts and opinions", variable=expressing_feelings, value="A", bg='#fffff6').pack(anchor='w')
        Radiobutton(questionsSet3_frame, text="B) Your emotions and feelings", variable=expressing_feelings, value="B", bg='#fffff6').pack(anchor='w')

        Label(questionsSet3_frame, text="Do you prefer:", font=("Arial Bold", 18), bg='#fffff6').pack(pady=10, anchor='w')
        Radiobutton(questionsSet3_frame, text="A) Having a structured and organized lifestyle", variable=structured_lifestyle, value="A", bg='#fffff6').pack(anchor='w')
        Radiobutton(questionsSet3_frame, text="B) Being spontaneous and adaptable", variable=structured_lifestyle, value="B", bg='#fffff6').pack(anchor='w')

        Label(questionsSet3_frame, text="Are you more comfortable:", font=("Arial Bold", 18), bg='#fffff6').pack(pady=10, anchor='w')
        Radiobutton(questionsSet3_frame, text="A) Making decisions quickly and sticking to plans", variable=quick_decisions_options, value="A", bg='#fffff6').pack(anchor='w')
        Radiobutton(questionsSet3_frame, text="B) Keeping your options open and being flexible", variable=quick_decisions_options, value="B", bg='#fffff6').pack(anchor='w')

        Label(questionsSet3_frame, text="Do you prefer:", font=("Arial Bold", 18), bg='#fffff6').pack(pady=10, anchor='w')
        Radiobutton(questionsSet3_frame, text="A) Having closure and resolution", variable=closure_possibilities, value="A", bg='#fffff6').pack(anchor='w')
        Radiobutton(questionsSet3_frame, text="B) Exploring possibilities and staying open-ended", variable=closure_possibilities, value="B", bg='#fffff6').pack(anchor='w')





        # Submit button
        submit_btn = Button(questionsSet3_frame, text="Submit", command=collect_answers)
        submit_btn.pack(pady=20)
        next_btn1 = Button(questionsSet1_frame, text="Next", command=show_questions_set2)
        next_btn1.pack(pady=20)
        next_btn2 = Button(questionsSet2_frame, text="Next", command=show_questions_set3)
        next_btn2.pack(pady=20)

        # Label to display results

      

        # Initially hide questions frame
        questionsSet1_frame.pack_forget()
        questionsSet2_frame.pack_forget()
        questionsSet3_frame.pack_forget()
        results_frame.pack_forget()

        self.form.mainloop()

            
        yield Fact(action="find_personality")                   
         
        
            # print("Welcome to the Personality Type Expert System!")
            # print("Please answer the following questions to determine your personality type.")
            # yield Fact(action="find_personality")
            

        #set 1

    @Rule(Fact(action='find_personality'), NOT(Fact(extraverted=W())))
    def question_extraverted(self):
        #print("Extraversion (E) vs. Introversion (I):")
        # print("When do you enjoy the most:")
        # print("A) Social gatherings and being around people")
        # print("B) You prefer to spend quiet time alone or with a few close friends.")
        self.declare(Fact(extraverted=self.answers[0]))

    @Rule(Fact(action='find_personality'), NOT(Fact(social_activity=W())))
    def question_social_activity(self):
        # print("After a long day, Would you rather:")
        # print("A)Go to a party")
        # print("B)stay home and read a book")
        self.declare(Fact(social_activity=self.answers[1]))

    @Rule(Fact(action='find_personality'), NOT(Fact(energized=W())))
    def question_energized(self):
        # print("When do you feel energized:")
        # print("A)By interacting with others")
        # print("B)By spending time alone")
        self.declare(Fact(energized=self.answers[2]))
        
        #set2

    @Rule(Fact(action='find_personality'), NOT(Fact(facts_details=W())))
    def question_facts_details(self):

        self.declare(Fact(facts_details=self.answers[3]))

    @Rule(Fact(action='find_personality'), NOT(Fact(concrete_abstract=W())))
    def question_concrete_abstract(self):
       
        self.declare(Fact(concrete_abstract=self.answers[4]))

    @Rule(Fact(action='find_personality'), NOT(Fact(experience_imagination=W())))
    def question_experience_imagination(self):
        # print("When solving problems, do you rely more on:")
        # print("A)your past experiences")
        # print("B)your imagination")
        self.declare(Fact(experience_imagination=self.answers[5]))
        
        #set3

    @Rule(Fact(action='find_personality'), NOT(Fact(logic_emotions=W())))
    def question_logic_emotions(self):
        # print("Are you more likely to make decisions based on:")
        # print("A)logic and reason")
        # print("B)values and emotions")
        self.declare(Fact(logic_emotions=self.answers[6]))

    @Rule(Fact(action='find_personality'), NOT(Fact(fairness_efficiency=W())))
    def question_fairness_efficiency(self):
        # print("Do you prioritize in decision-making:")
        # print("A)fairness and justice")
        # print("B)efficiency and objectivity")
        self.declare(Fact(fairness_efficiency=self.answers[7]))

    @Rule(Fact(action='find_personality'), NOT(Fact(expressing_feelings=W())))
    def question_expressing_feelings(self):
        # print("Are you more comfortable expressing:")
        # print("A)your thoughts and opinions")
        # print("B)your emotions and feelings")
        self.declare(Fact(expressing_feelings=self.answers[8]))
        
        #set4

    @Rule(Fact(action='find_personality'), NOT(Fact(structured_lifestyle=W())))
    def question_structured_lifestyle(self):
        # print("Do you prefer:")
        # print("A)Having a structured and organized lifestyle")
        # print("B)Being spontaneous and adaptable")
        self.declare(Fact(structured_lifestyle=self.answers[9]))

    @Rule(Fact(action='find_personality'), NOT(Fact(quick_decisions_options=W())))
    def question_quick_decisions_options(self):
        # print("Are you more comfortable:")
        # print("A)making decisions quickly and sticking to plans")
        # print("B)keeping your options open and being flexible")
        self.declare(Fact(quick_decisions_options=self.answers[10]))

    @Rule(Fact(action='find_personality'), NOT(Fact(closure_possibilities=W())))
    def question_closure_possibilities(self):
        # print("Do you prefer:")
        # print("A)having closure and resolution")
        # print("B)exploring possibilities and staying open-ended")
        self.declare(Fact(closure_possibilities=self.answers[11]
                          ))



###################################################################################################3

    
  

    @Rule(Fact(action='find_personality'),
          Fact(extraverted=MATCH.extraverted),
          Fact(social_activity=MATCH.social_activity),
          Fact(energized=MATCH.energized),
          Fact(facts_details=MATCH.facts_details),
          Fact(concrete_abstract=MATCH.concrete_abstract),
          Fact(experience_imagination=MATCH.experience_imagination),
          Fact(logic_emotions=MATCH.logic_emotions),
          Fact(fairness_efficiency=MATCH.fairness_efficiency),
          Fact(expressing_feelings=MATCH.expressing_feelings),
          Fact(structured_lifestyle=MATCH.structured_lifestyle),
          Fact(quick_decisions_options=MATCH.quick_decisions_options),
          Fact(closure_possibilities=MATCH.closure_possibilities))


    def determine_personality(self, extraverted, social_activity, energized,
                              facts_details, concrete_abstract, experience_imagination,
                              logic_emotions, fairness_efficiency, expressing_feelings,
                              structured_lifestyle, quick_decisions_options, closure_possibilities):
      
        # Calculate the majority for each set of questions
        majority_1 = self.calculate_majority([
            ('extraverted', extraverted),
            ('social_activity', social_activity),
            ('energized', energized)
        ])
        majority_2 = self.calculate_majority([
            ('facts_details', facts_details),
            ('concrete_abstract', concrete_abstract),
            ('experience_imagination', experience_imagination)
        ])
        majority_3 = self.calculate_majority([
            ('logic_emotions', logic_emotions),
            ('fairness_efficiency', fairness_efficiency),
            ('expressing_feelings', expressing_feelings)
        ])
        majority_4 = self.calculate_majority([
            ('structured_lifestyle', structured_lifestyle),
            ('quick_decisions_options', quick_decisions_options),
            ('closure_possibilities', closure_possibilities)
        ])
     
        # Determine personality based on majorities
        if majority_1 == 'A' and majority_2 == 'A' and majority_3 == 'A' and majority_4 == 'A':
            self.declare(Fact(personality="ESTJ"))
        elif majority_1 == 'A' and majority_2 == 'A' and majority_3 == 'A' and majority_4 == 'B':
            self.declare(Fact(personality="ESTP"))
        elif majority_1 == 'A' and majority_2 == 'A' and majority_3 == 'B' and majority_4 == 'A':
            self.declare(Fact(personality="ESFJ"))
        elif majority_1 == 'A' and majority_2 == 'A' and majority_3 == 'B' and majority_4 == 'B':
            self.declare(Fact(personality="ESFP"))
        elif majority_1 == 'A' and majority_2 == 'B' and majority_3 == 'A' and majority_4 == 'A':
            self.declare(Fact(personality="ENTJ"))
        elif majority_1 == 'A' and majority_2 == 'B' and majority_3 == 'A' and majority_4 == 'B':
            self.declare(Fact(personality="ENTP"))
        elif majority_1 == 'A' and majority_2 == 'B' and majority_3 == 'B' and majority_4 == 'A':
            self.declare(Fact(personality="ENFJ"))
        elif majority_1 == 'A' and majority_2 == 'B' and majority_3 == 'B' and majority_4 == 'B':
            self.declare(Fact(personality="ENFP"))
        elif majority_1 == 'B' and majority_2 == 'A' and majority_3 == 'A' and majority_4 == 'A':
            self.declare(Fact(personality="ISTJ"))
        elif majority_1 == 'B' and majority_2 == 'A' and majority_3 == 'A' and majority_4 == 'B':
            self.declare(Fact(personality="ISTP"))
        elif majority_1 == 'B' and majority_2 == 'A' and majority_3 == 'B' and majority_4 == 'A':
            self.declare(Fact(personality="ISFJ"))
        elif majority_1 == 'B' and majority_2 == 'A' and majority_3 == 'B' and majority_4 == 'B':
            self.declare(Fact(personality="ISFP"))
        elif majority_1 == 'B' and majority_2 == 'B' and majority_3 == 'A' and majority_4 == 'A':
            self.declare(Fact(personality="INTJ"))
        elif majority_1 == 'B' and majority_2 == 'B' and majority_3 == 'A' and majority_4 == 'B':
            self.declare(Fact(personality="INTP"))
        elif majority_1 == 'B' and majority_2 == 'B' and majority_3 == 'B' and majority_4 == 'A':
            self.declare(Fact(personality="INFJ"))
        elif majority_1 == 'B' and majority_2 == 'B' and majority_3 == 'B' and majority_4 == 'B':
            self.declare(Fact(personality="INFP"))

    @Rule(Fact(action='find_personality'), Fact(personality=MATCH.personality))
    def print_personality(self, personality):
        print("Your Myers-Briggs Type Indicator (MBTI) personality type is: %s" % personality)
        self.personality = personality
     
        resultform = Tk()
        resultform.title("Personality Type Expert System")
        resultform.geometry("800x600")
        resultform.configure(bg='#fffff6')
        header_label = Label(resultform, text="Personality Type Expert System", font=("Arial Bold", 16), bg='black', fg='white', pady=14, padx=14, anchor='w')
        header_label.pack(fill='x')
        results_frame = Frame(resultform, bg='#fffff6')
        results_frame.pack(expand=True)
        Label(results_frame, text="Your Myers-Briggs Type Indicator (MBTI) personality type is: %s" % personality, font=("Arial Bold", 18), bg='#fffff6').pack(pady=20)
        typephoto = PhotoImage(file="assets/%s.png" % personality)
        typephoto_label = Label(results_frame, image=typephoto, width=350, height=350)  # Adjust width and height according to your desired dimensions
        typephoto_label.image = typephoto
        typephoto_label.pack(pady=20)
        descriptions = {
            "ESTJ": "Great organizers, realistic and systematic.",
            "ESTP": "Flexible, tolerant, takes a pragmatic approach.",
            "ESFJ": "Warm-hearted, cooperative, and responsible.",
            "ESFP": "Outgoing, friendly, and accepting.",
            "ENTJ": "Bold, imaginative, and strong-willed leaders.",
            "ENTP": "Quick, ingenious, and alert; can be outspoken.",
            "ENFJ": "Charismatic, inspiring, and able to mesmerize their listeners.",
            "ENFP": "Enthusiastically creative and sociable free spirits.",
            "ISTJ": "Practical, fact-minded, and reliable.",
            "ISTP": "Tolerant, flexible, quiet observers until a problem appears.",
            "ISFJ": "Very dedicated, warm protectors, always ready to defend.",
            "ISFP": "Artistic, sensitive, and playful.",
            "INTJ": "Imaginative and strategic thinkers with a plan for everything.",
            "INTP": "Innovative inventors with an unquenchable thirst for knowledge.",
            "INFJ": "Quiet and mystical, yet very inspiring and tireless idealists.",
            "INFP": "Poetic, kind, and altruistic people, always eager to help."
        }
        description_label = Label(results_frame, text=descriptions[personality], font=("Arial Bold", 14), bg='#fffff6')
        description_label.pack(pady=20)




        resultform.mainloop()
        
    
        

        
        
        

engine = PersonalityExpert()
engine.reset()  # Prepare the engine for the execution.
engine._initial_action()  # Call this before starting the rule engine loop.
engine.run()  # Start the rule engine loop.

```
